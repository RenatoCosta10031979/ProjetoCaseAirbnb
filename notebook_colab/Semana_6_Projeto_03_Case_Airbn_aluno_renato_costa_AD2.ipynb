```python
#Neste bloco de código será verificado se o DataFrame apresenta colunas repetidas(iguais)
#Será utilizar a estrutura de repetição 'For' e a estrutura condicional 'if e else',
# e o método tolist() para converter as colunas em estrutura de lista para que possamos verificar as colunas
# Obter a lista de nomes das colunas do DataFrame
colunas = df.columns

# Criar um conjunto vazio para armazenar as colunas duplicadas
# O set() é usado para dizer ao python armazenar os elementos repetidos (caso existir), na variável 'colunas_duplicadas'
colunas_duplicadas = set()

# Loop for para verificar colunas duplicadas
#O count() verifica a quantidade de vezes ( ocorrência) que a coluna
# que foi passado dentro do parâmetro (count(coluna))
for coluna in colunas:
    if colunas.tolist().count(coluna) > 1:
        colunas_duplicadas.add(coluna)

# Verificar se há colunas duplicadas e exibir os resultados
if len(colunas_duplicadas) > 0:
    print("As seguintes colunas estão duplicadas:")
    for coluna_duplicada in colunas_duplicadas:
        print(coluna_duplicada)
else:
    print("Não há colunas duplicadas no DataFrame.")
```

Aplicando o comando **unique()** para verificar os valores únicos e excluir possíves valores duplicados. Será executado para cada coluna escolhida para análise.

```python
#A estrutura de repetição 'for' permite iterar cada coluna ao utilizado o comando df.columns
# o df.columns acessa o nome de cada coluna do DataFrame
for coluna in df.columns:
   valores_unicos = df[coluna].unique() # df[coluna].unique() varre cada coluna e verifica se há valor único e excluir as duplicadas caso existir e armazena na variável valores_únicos
# Ao executar o comando display com o parâmetro valores_unicos verificamos os valores únicos
#Observação: Não conseguir armazenar os valores em um novo formato de Dataframe. Os dados ficaram
#dispersos. Contudo, o objetivo principal foi atingido que é assegurar os valores únicos e excluir possíveis valores duplicados nas colunas.
   display(valores_unicos)

```

### **Qualidade dos Dados**

A **validação de esquema** é usada para definir tipos de dados e restrições em um DataFrame para garantir a integridade dos dados no pandas.

Ao chamar **schema.validate(df)**, o DataFrame df é validado em relação ao esquema definido. Cada coluna é verificada em relação às regras e restrições especificadas no esquema. Se alguma coluna não atender às regras, uma exceção será lançada, indicando o motivo da falha na validação.

### **Combinando filtro e agrupamento**


Os próximos dois blocos de códigos abaixo realizar  uma filtragem considerando que o proprietário possuí a classficação como superhost
 E foi fixado alguns valores para que depois seja comparado com mesmo código porém com a classificação do prietário não
 é do tipo superhost. O intuito é verificar se quando estabelecimentos que estão na mesma região de distância do centro
 e distância do metro e com a mesma classificação de limpeza e satistação geral se o fato do proprietário ser
ou não superhost afeta os valores dos preço total. E essa filtragem está dividida em Analise 1 e Análise 2

Análise 1

```python
#Filtrando o DataFrame 'df_analise' com base em critérios específicos escolhidos pelo analista
# Neste código foi feito uma filtragem considerando que o proprietário possuí a classficação como superhost
# E foi fixado alguns valores para que depois seja comparado com mesmo código porém com a classificação do prietário não
# é do tipo superhost. O intuito é verificar se quando estabelecimentos que estão na mesma região de distância do centro
# e distância do metro e com a mesma classificação de limpeza e satistação geral se o fato do proprietário ser
#ou não superhost afeta os valores dos preço total. E essa filtragem está dividida em Analise 1 e Análise 2
filtro_qte_quartos = (df_analise['qte_quartos'] == 3)
filtro_capacidade_pessoas = (df_analise['capacidade_de_pessoas'] == 4)
filtro_tipo_quarto = df_analise['tipo_de_quarto'] == 'Entire home/apt'
filtro_distancia_metro = (df_analise['distancia_metro'] <= 1 ) # quanto mais longe a hospedagem fica da estação de metrô mais barato fica a hospedagem
filtro_distancia_centro = (df_analise['distancia_centro'] < 1.5)# & (df_analise['distancia_centro'] <= 5) # a unidade de medida é km.
filtro_anfitriao_superhost = df_analise['anfitriao_superhost'] == True
filtro_classificacao_limpeza = df_analise['classificacao_limpeza'] >= 7
filtro_satisfacao_geral_hospede = df_analise['satisfacao_geral_hospede'] >= 70

# Abaixo temos a aplicação de todos os filtros acima mencionados ao DataFrame 'df_analise' usando a função 'loc'.
df_filtrado = df_analise.loc[
    filtro_qte_quartos &
    filtro_capacidade_pessoas &
    filtro_distancia_metro &
    filtro_distancia_centro &
    filtro_anfitriao_superhost &
    filtro_classificacao_limpeza &
    filtro_satisfacao_geral_hospede
]

# O código abaixo vai agrupar as linhas do DataFrame 'df_analise' pelo
#pelo valor da coluna 'tipo_de_quarto' e calcular a média dos valores da coluna
#'preco_total' para cada grupo especificado na filtragem.
media_preco_total = df_filtrado.groupby('tipo_de_quarto')['preco_total'].mean()

# Imprimir a média do 'preco_total' obtida após a aplicação dos filtros e agrupamento.
display(media_preco_total)
```

```python
# Nesta epata iremos utilizar uma  serviço de geocodificação reversa
# permitindo obter o endereço a partir das coordenadas geográficas.
#Para localização de endereço foi utilizado o serviço: MapQuest Reverse Geocoding API.
#O link para acessar o MapQuest está disponível no ínicio deste projeto no tópico: Referência
import requests

latitude = 52.4
longitude = 4.895

url = f'https://nominatim.openstreetmap.org/reverse?format=jsonv2&lat={latitude}&lon={longitude}'
response = requests.get(url)
data = response.json()

endereco = data['display_name']

print(f'Endereço: {endereco}')
```
